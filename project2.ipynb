# Heart Disease

```python
#PROJECT 2(Heart Disease)
# Import libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load the dataset
file_path = '/content/drive/MyDrive/Data mining/heart_disease.csv'
df = pd.read_csv(file_path)
print("Initial Dataset:")
print(df.head())

# Step 1: Data Preprocessing
# Convert all categorical columns to dummy/one-hot encoded variables
df_encoded = pd.get_dummies(df, columns=df.columns)
print("Encoded Dataset:")
print(df_encoded.head())

# Step 2: Apriori Algorithm
# Generate frequent itemsets using Apriori
frequent_itemsets = apriori(df_encoded, min_support=0.3, use_colnames=True)
print("Frequent Itemsets (Support >= 0.3):")
print(frequent_itemsets)

# Step 3: Association Rules
# Generate association rules using confidence = 0.7
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Sort rules by confidence, then lift, then support
rules = rules.sort_values(by=['confidence', 'lift', 'support'], ascending=False)

# Display Top 10 Rules
top_10_rules = rules.head(10)
print("Top 10 Association Rules:")
print(top_10_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# Step 4: Rule Interpretation
# Select one rule to interpret
selected_rule = top_10_rules.iloc[0]
antecedent = list(selected_rule['antecedents'])
consequent = list(selected_rule['consequents'])
confidence = selected_rule['confidence']
lift = selected_rule['lift']
support = selected_rule['support']

print("\n--- Selected Rule Explanation ---")
print(f"If a record has: {antecedent}")
print(f"Then it is likely to have: {consequent}")
print(f" - Support: {support:.2f}")
print(f" - Confidence: {confidence:.2f}")
print(f" - Lift: {lift:.2f}")
```

# Mushrooms

```python
#pPROJECT 2(Mushrooms)
# Importing necessary libraries
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from google.colab import drive

# Mounting Google Drive
drive.mount('/content/drive')

# Load the Mushroom Dataset
file_path = '/content/drive/MyDrive/Data mining/mushrooms.csv'
df = pd.read_csv(file_path)

# Display first few rows
df.head()

# Step 1: Preprocessing - One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=df.columns)
df_encoded.head()

# Step 2: Frequent Itemset Generation using Apriori
frequent_itemsets = apriori(df_encoded, min_support=0.3, use_colnames=True)
print("Frequent Itemsets (Support >= 0.3):")
print(frequent_itemsets)

# Step 3: Generate Association Rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules = rules.sort_values(by=['confidence', 'lift', 'support'], ascending=False)

# Display top 10 rules
top_rules = rules.head(10)
print("Top 10 Association Rules:")
print(top_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# Step 4: Interpret one selected rule
# You can choose any of the top_rules to analyze
selected_rule = top_rules.iloc[0]
print("\nSelected Rule Analysis:")
print(f"If {selected_rule['antecedents']} then {selected_rule['consequents']}")
print(f"Support: {selected_rule['support']}")
print(f"Confidence: {selected_rule['confidence']}")
print(f"Lift: {selected_rule['lift']}")
```
